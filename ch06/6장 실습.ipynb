## 6.2 가중치의 초기값

### 6.2.2 은닉층의 활성화값 분표

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.random.randn(1000, 100)
node_num = 100
hidden_layer_size = 5
activations = {}

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i - 1]
    
    w = np.random.randn(node_num, node_num) / np.sqrt(node_num)
    a = np.dot(x, w)
    z = sigmoid(a)
    activations[i] = z
```

```python
for i, a in activations.items():
    plt.subplot(1, len(activations), i + 1)
    plt.title(str(i + 1) + '-layer')
    plt.hist(a.flatten(), 30, range=(0, 1))
plt.show()
```

```python
activations[4][0]
```

### 6.2.3. ReLU 가중치 초깃값

```python
def relu(x):
    return np.maximum(0, x)

x = np.random.randn(1000, 100)
node_num = 100
hidden_layer_size = 5
activations = {}

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i - 1]
    
    w = np.random.randn(node_num, node_num) * np.sqrt(2 / node_num)
    a = np.dot(x, w)
    z = relu(a)
    activations[i] = z
```

```python
for i, a in activations.items():
    plt.subplot(1, len(activations), i + 1)
    plt.title(str(i + 1) + '-layer')
    plt.hist(a.flatten(), 30, range=(0, 1))
plt.show()
```

```python
activations[4][0]
```

결과: 넓게 퍼지긴 하는데 여러번 곱할수록 점점 값이 커짐.

```python
def relu(x):
    return np.maximum(0, x)

x = np.random.randn(1000, 100)
node_num = 100
hidden_layer_size = 5
activations = {}

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i - 1]
    
    w = np.random.randn(node_num, node_num) + 0.001
    a = np.dot(x, w)
    z = relu(a)
    activations[i] = z
```

```python
for i, a in activations.items():
    plt.subplot(1, len(activations), i + 1)
    plt.title(str(i + 1) + '-layer')
    plt.hist(a.flatten(), 50, range=(0, 100))
plt.show()
```
